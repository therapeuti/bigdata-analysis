### 1-5. Iris 데이터 세트에서 sepal_length 변수가 이상치를 가지는 데이터 행 수를 계산하세요. [단, 이상치 범위는 (평균 + 2 × 표준편차) 이상이거나 (평균 – 2 × 표준편차) 이하이다.]


```python
# prompt: average_list 값들의 표준편차, 평균 계산

import numpy as np

# average_list의 표준편차 계산
std_dev = np.std(average_list)
print("표준편차:", std_dev)

# average_list의 평균 계산
mean_val = np.mean(average_list)
print("평균:", mean_val)
```

```python
#sepal length 이상치 행 수 계산. 이상치 범위는 평균 + 2 * 표준편차 이상이거나 평균 - 2* 표준편차 이하

sep_avg = df['sepal_length'].mean() # 평균
sep_std = df['sepal_length'].std() # 표준편차

outlier = (df['sepal_length'] >= sep_avg + 2*sep_std) | (df['sepal_length'] <= sep_avg - 2*sep_std) # 이상치 범위 설정

len(df.loc[outlier, 'sepal_length']) # 행 개수 구하기
```

# 2. Wine 데이터


```python
import pandas as pd

df = pd.read_csv('/content/yemoonsaBigdata/datasets/Part2/wine.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Ai가 써준 코드

# total_phenols 결측치 제거하고, alcalinity_of_ash 결측치는 중앙값으로 대체 후 alcalinity_of_ash 평균 계산
df_dropna = df.dropna(subset=['total_phenols'])
median_alcalinity_of_ash = df_dropna['alcalinity_of_ash'].median()
df_dropna['alcalinity_of_ash'] = df_dropna['alcalinity_of_ash'].fillna(median_alcalinity_of_ash)
result1 = int(df_dropna['alcalinity_of_ash'].mean())

# alcohol 값이 평균 이상이고 color_intensity 값이 평균 이상인 데이터만 남겼을 때 가장 많은 class 값
alcohol_mean = df['alcohol'].mean()
color_intensity_mean = df['color_intensity'].mean()
df_filtered = df[(df['alcohol'] >= alcohol_mean) & (df['color_intensity'] >= color_intensity_mean)]
result2 = df_filtered['class'].value_counts().idxmax()

# class 1에 해당하는 데이터의 ash 평균값과 모든 결측치 제거 후 ash 평균값의 차이의 절댓값
df_class1 = df[df['class'] == 1]
ash_mean_class1 = df_class1['ash'].mean()
df_dropna_all = df.dropna()
ash_mean_dropna = df_dropna_all['ash'].mean()
result3 = round(abs(ash_mean_class1 - ash_mean_dropna), 3)

# alcohol 변수의 상위 20번째 값으로 상위 20개 값 변환 후, proanthocyanins가 2 이하인 데이터의 alcohol 평균
alcohol_sorted = sorted(df['alcohol'])
twentieth_value = alcohol_sorted[19]
df_top20 = df.copy()
df_top20['alcohol'] = df_top20['alcohol'].apply(lambda x: twentieth_value if x >= twentieth_value else x)
df_filtered4 = df_top20[df_top20['proanthocyanins'] <= 2]
result4 = round(df_filtered4['alcohol'].mean(), 2)

# class 1 proline 3사분위수와 1사분위수 차, class 2 proline 3사분위수와 1사분위수 차, 두 값 차이의 절댓값
df_class1 = df[df['class'] == 1]
q3_class1 = df_class1['proline'].quantile(0.75)
q1_class1 = df_class1['proline'].quantile(0.25)
diff_class1 = q3_class1 - q1_class1
df_class2 = df[df['class'] == 2]
q3_class2 = df_class2['proline'].quantile(0.75)
q1_class2 = df_class2['proline'].quantile(0.25)
diff_class2 = q3_class2 - q1_class2
result5 = abs(diff_class1 - diff_class2)

print(result1)
print(result2)
print(result3)
print(result4)
result5
```

### 2-5. Wine 데이터 세트에서 class 1에 해당하는 데이터의 proline 변수의 3사분위수와 제1사분위수 차를 구하고, class 2에 해당하는 데이터의 proline 변수의 3사분위수와 제1사분위수 차를 구하여 두 값의 차이의 절댓값을 계산하시오.

```python
# Wine 데이터 세트에서 class 1에 해당하는 데이터의 proline 변수의 3사분위수와 제1사분위수 차를 구하고,
# class 2에 해당하는 데이터의 proline 변수의 3사분위수와 제1사분위수 차를 구하여 두 값의 차이의 절댓값을 계산하시오.

q75 = df[df['class']==1]['proline'].quantile(0.75)  #class1에 해당하는 proline 변수의 3사분위수 구하기
q25 = df[df['class']==1]['proline'].quantile(0.25)  #class1에 해당하는 proline 변수의 1사분위수 구하기

dff1 = q75 - q25

cl2_75 = df[df['class']==2]['proline'].quantile(0.75)   #class2에 해당하는 proline 변수의 3사분위수 구하기
cl2_25 = df[df['class']==2]['proline'].quantile(0.25)   #class2에 해당하는 proline 변수의 1사분위수 구하기

dff2 = cl2_75 - cl2_25

result = abs(dff1 - dff2)  #abs로 절대값 구하기
result
```
